# Methods on the Groupby Object and DataFrame Columns

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector") 
#this is a sectors groupby object i.e. grouping or segment of the dataframe
#we need to start calling methods on a sectors (groupby object) to actually make use of it
fortune.head(3)
```

```python
sectors.max()
#represents the last compnay represented alphabetically
```

```python
sectors.min()
#the first company that begins in the alphabet in the grouping will appear
```

```python
sectors.sum()
#takes the sum of all available columns and displays all those columns you can perform a sum on
```

```python
#giving avg. of Revenues, Profits and Employees col. in each grouping
sectors.mean()
```

```python
sectors["Revenue"].sum()
#if you wanna just get the sum of values only in 1 column i.e. Revenues in this case
```

```python
#total no. of employees per sector
sectors["Employees"].sum()
```

```python
# best performing profit value in that sector 
sectors["Profits"].max()
```

```python
#worst performing profit values 
sectors["Profits"].min()
```

```python
sectors["Profits", "Revenue"].sum()
#a dataframe where we are extracting sum total of revenues and profits in that particular grouop
```

# The .agg( ) Method

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector") 

fortune.head(3)
```

```python
sectors.agg({"Revenue": "sum",
            "Profits" : "sum",
            "Employees": "mean"
            })
#expects a python dictionary where the key is the column name and the value is the 
#method we wanna apply in this case sum
```

```python
#to apply multiple operations to multiple columns can pass it as a list
sectors.agg(["size", "sum", "mean"])
```

```python
sectors.agg({"Revenue":["sum", "mean"],
            "Profits": "sum",
            "Employees" : "mean"
            })
```

# Iterating Through Groups

```python
fortune = pd.read_csv("fortune1000.csv", index_col = "Rank")
sectors = fortune.groupby("Sector") 
fortune.head(3)
```

```python
#get the company with the max profit in each sector

df = pd.DataFrame(columns = fortune.columns)
df
```

```python
for sector, data in sectors:
    highest_revenue_company_in_group = data.nlargest(1, "Revenue")
    df = df.append(highest_revenue_company_in_group)
```

```python
df
```

```python
#we wanna find highest revenue in a city based analysis
cities = fortune.groupby("Location") # we wanna groupby location
df = pd.DataFrame(columns = fortune.columns)
df
#ready to iterate over it and append to it
```

```python
#gotaa give a temporary variable to represent each grouping 
#then give a second temporary variable to represent each actual result 

for city, data in cities: 
    highest_revenue_in_city = data.nlargest(1, "Revenue") #getting the top performing result in the Revenues column
    df = df.append(highest_revenue_in_city)
```
