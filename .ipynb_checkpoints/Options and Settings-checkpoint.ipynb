# Changing pandas Options with Methods

```python
data = np.random.randint(0,100,[1000,50]) #assign to data variable
df = pd.DataFrame(data)
df.tail(2)
```

```python
pd.get_option("max_rows")
```

```python
pd.get_option("max_columns")
```

```python
pd.set_option("max_rows", 16)
```

```python
df
```
