# Bar Charts

```python
google = data.DataReader(name = "GOOG", data_source = "yahoo", start = "2004-01-01", end = "2018-12-31")
google.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <= 200:
        return "Poor"
    elif stock_price >= 200 and stock_price <= 500:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
google["Close"]
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "bar")
#the default parameter of plot has something called kind which is set to line, hence we see line graph
#we can update it to the bar to get a bar graph
```

```python
google["Close"].apply(rank_performance).value_counts().plot(kind = "barh")
```

# Pie Charts

```python
apple = data.DataReader(name = "AAPL", data_source= "yahoo", start="2012-01-01", end="2019-03-31")
apple.head(3)
```

```python
apple["Close"].mean() #giving us the avg of apple stock price
```

```python
def rank_performance (stock_price):
    if stock_price >= 116.99025965548688:
        return "Above Average"
    else:
        return "Below Average"
```

```python
apple["Close"].apply(rank_performance).value_counts().plot(kind = "pie", legend = True)
```

# Histograms

```python
google = data.DataReader(name = "GOOG", data_source = "yahoo", start = "2004-01-01", end = "2018-12-31")
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
custom_round(560.2)
```

```python
google["High"].apply(custom_round).value_counts().sort_index()
```

```python
google["High"].apply(custom_round).nunique()
```

```python
google["High"].apply(custom_round).plot(kind = "hist", bins = 13)
```
