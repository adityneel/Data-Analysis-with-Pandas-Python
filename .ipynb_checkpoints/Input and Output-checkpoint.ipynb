```python
import pandas as pd
```

# Feed pd.read_csv( ) Method a URL Argument

```python
URL = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names = pd.read_csv(URL)
```

```python
baby_names.head()
```

# Quick Object Conversions

```python
baby_names["Child's First Name"].tolist()

baby_names["Child's First Name"]
```

```python
#converting series to a dataframe
baby_names["Child's First Name"].to_frame()
```

```python
#to extract a string of all the names in the Child first name column
#name is the temporary variable that I will be assigning to the Child first name column
#using the join method to perform a particular operation 

#what this operation does is that for every name in baby_names DF , use str() to check whether they are
#strings and if yes then join with a ,
", ".join(str(name) for name in baby_names["Child's First Name"])
```

# Export CSV File with the .to_csv( ) Method

```python
baby_names.to_csv("NYC Baby Names.csv", index = False, columns = ["Year of Birth","Child's First Name","Count"], encoding = "utf-8")
#exported the csv to current working directory i.e. where we are working in the pandas folder on desktop
#utf-8 includes all the characters of modern languages . if any encoding issues =, always use this param
```

# Export Excel File

```python
URL = "https://data.cityofnewyork.us/api/views/25th-nujf/rows.csv"
baby_names = pd.read_csv(URL)
baby_names.head(3)
```

```python
girls = baby_names[baby_names["Gender"] == "FEMALE"]
boys = baby_names[baby_names["Gender"] == "MALE"]
```

```python
excel_file = pd.ExcelWriter("Baby Names.xlsx")
```

```python
girls.to_excel(excel_file, sheet_name = "Girls", index = False)
boys.to_excel(excel_file, sheet_name = "Boys", index = False, columns = ["Gender", "Child's First Name", "Count"])
```

```python
excel_file.save()
```
