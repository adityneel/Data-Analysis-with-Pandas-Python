# Filtering with String Methods

```python
chicago = pd.read_csv("chicago.csv").dropna(how = "all") 
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago["Position Title"].str.lower()
```

```python
chicago["Position Title"].str.lower().str.contains("water")
#contains() takes a single argument of the string you are trying to search 
#returns TRUE if able to find the substring else returns false
```

```python
#now that we have a boolean series as seen above, we are al set to filter those position titles that have water in them 
mask = chicago["Position Title"].str.lower().str.contains("water")
chicago[mask]
```

```python
chicago["Position Title"].str.lower().str.startswith("water")
```

```python
chicago[chicago["Position Title"].str.lower().str.startswith("water")]
```

```python
mask1 = chicago["Position Title"].str.lower().str.endswith("ist")
chicago[mask1]
```

# The expand and n Paramters of the str.split( ) Method

```python
chicago = pd.read_csv("chicago.csv").dropna(how = "all") 
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago["Name"].str.split(",", expand = True)
#expand by defualt set to False
#when we set expand = True, it is gonna return a dataframe instead of a list, where the list values
#will be placed across each col of the dataframe
```

```python
chicago[["First Name", "Last Name"]] = chicago["Name"].str.split(",", expand = True)
#so its gonna use the index labels from our original dataframe and assign it to these labels in the liost
#that we are passing in chicago on the right side
```

```python
chicago.head(3)
```

```python
chicago["Position Title"].str.split(" ", expand = True, n = 1)
#n represents the max no. of splits we want, so after the first space it will put all the remaining in 2nd column
```

```python
chicago[["First TitLe Word", "Remaining Words"]] = chicago["Position Title"].str.split(" ", expand = True, n = 1)
```
