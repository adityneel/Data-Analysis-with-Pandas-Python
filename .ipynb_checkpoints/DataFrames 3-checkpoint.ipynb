# The .nsmallest( ) and .nlargest( ) Methods 

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
#normally we can use the sort_values() method
#for instance to get the top 3 grossing films we can write as follows
bond.sort_values("Box Office", ascending = False).head(3)
```

```python
#nlargest does the same as the above 
#this can be more efficient rather than sorting entire dataframe
bond.nlargest(3, columns = "Box Office")
```

```python
#getting the 2 films with least Box office
bond.nsmallest(n = 2, columns = "Box Office")
```

```python
#get the 3 largest budget films
bond.nlargest(3,columns = "Budget")
```

```python
#get the 6 smallest bond actor salaries
bond.nsmallest(6, columns = "Bond Actor Salary")
```

```python
#if you wanna extract only the sorted Box office films can do the following by calling the Box Office series
#and specifying the nlargest() parameter
bond["Box Office"].nlargest(8)
```

```python
bond["Year"]
```

```python
#get the first 2 bond films made
bond["Year"].nsmallest(2)
```

# Filtering with the where Method

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond["Actor"] == "Sean Connery"
```

```python
mask = bond["Actor"] == "Sean Connery"
bond[mask]
```

```python
#where() method accepts a boolean series
bond.where(mask)
```

```python
bond["Box Office"] > 800
```

```python
bond.where(bond["Box Office"] > 800)
#only where the condition is met, those rows show up while others are nAN
```

```python
mask1 = bond["Actor"] == "Sean Connery"
mask2 = bond["Box Office"] > 800
```

```python
#we want thoise rows where box office > 800 and actor was Sean Connery
bond.where(mask1 & mask2)
```

# A Review of the .apply( ) Method on Single Columns

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + "MILLIONS"

bond["Box Office"] = bond["Box Office"].apply(convert_to_string_and_add_millions)
```

```python
bond["Budget"] = bond["Budget"].apply(convert_to_string_and_add_millions)
```

```python
#Instead of above we will write a for loop 
#create a temporary variable "col" that will iterate through each value in columns[] list
#and apply the string function to add MILLIONS

columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

# The .apply( ) Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row): #gonna pass each row in our dataframe
    
    actor = row[1] #take value in the actor col and iterate and assign value 
    budget = row[4]
    
    if actor == "Pierce Brosnon":
        return "The best"
    elif actor == "Roger Moore" and budget >40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis = "columns")

#we gotta give the axis parameter with argument = columns 
#think of it as moving across the columns for each row to feed the list of values 
#every row moving left to right across our 6 columns
```

# The .copy( ) Method

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
directors = bond["Director"] # assigning to a new variable
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mr. John Glen"
#referring to the series and providing the index label 
#and then trying to assign a new value to director
```

```python
directors.head(3)
```

```python
bond.head(3) # here also the value has got changed
```

```python
#re-execute the above csv code and nowe we will use the copy method to assign to a brand new series object
directors = bond["Director"].copy()
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mr John Glen"
```

```python
#now this is a copy. its a totally different object from the series in our bond dataframe 
directors.head(3)
```
