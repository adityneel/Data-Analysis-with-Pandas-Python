# Inner Joins, Part1

```python
#The key behind inner joins is that it represents the merging of commonly shared values
#the point / data that the 2 tables have in common is what is gonna be returned to us via the inner join
```

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head(2)
```

```python
week2.head(2)
```

```python
week1.merge(week2, how = "inner", on = "Customer ID")
#the parameter "on" only works when both have same col. names
```

```python
# PERFORM a fliteration process now to extract the values from the customer id column
week1[week1["Customer ID"] ==155]
```

```python
week2[week2["Customer ID"] ==155]
#Here we see 2 occurences of the same in week1 and week2
#so pandas is appending both and we have the duplicate values above
#just gonna merge the common customer id 
```

```python
week1.merge(week2, how = "inner", on = "Customer ID", suffixes = [" - Week1", " - Week2"])
```

# Outer Joins

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week1.head(2)
```

```python
week2.head(2)
```

```python
#merging 2nd dataframe into the 1st fully 
week1.merge(week2,how = "outer", on = "Customer ID", suffixes = [" -Week 1", " -Week 2"],
           indicator = True)
#customer id = 537 bought food id = 9 in 1st week and food id = 5 in the 2nd week
#whenever there is a NaN i.e. Null value that customer no. eg. 97 bought only in week1 AND not week2
#indicator parameter tells us which type of Join it was i.e. whether found in left DF or right DF or both
#thus a great indicator to tell us where it is coming from
```

```python
len (week1.merge(week2,how = "outer", on = "Customer ID", suffixes = [" -Week 1", " -Week 2"]))
#454 implies that there are 46 rows that had a valid food id in week 1 and in week 2
#whenever its unable to find a value it inserts NaN and then merges it 
```

```python
merged = week1.merge(week2,how = "outer", on = "Customer ID", suffixes = [" -Week 1", " -Week 2"],
           indicator = True)
merged["_merge"].value_counts() #calling value_Counts to see how the values are distributed
```

```python
merged["_merge"].isin(["left_only", "right_only"])
```

```python
mask = merged["_merge"].isin(["left_only", "right_only"])
merged[mask]
#getting every unique customer id from week 1 and from week2 but NOT BOTH
```

# The left_on and right_on Parameters

```python
week1 = pd.read_csv("Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("Restaurant - Customers.csv")
foods = pd.read_csv("Restaurant - Foods.csv")
```

```python
week2.head(3)
```

```python
customers.head(3)
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID")
#left_on rep the DF the merge method being called on i.e. week2 left dataframe
#right_on represents the customers df
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID").drop("ID", axis = "columns")
```

```python
week2.merge(customers, how = "left", left_on = "Customer ID", right_on = "ID", sort = True).drop("ID", axis = "columns")
```
