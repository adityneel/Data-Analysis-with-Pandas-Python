# Filter a DataFrame based on a condition

```python
df = pd.read_csv("employees.csv")
df["Start Date"] = pd.to_datetime(df["Start Date"])
df ["Last Login Time"] = pd.to_datetime(df["Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
df[df["Gender"] == "Male"] #doing comparision & only giving the rows with Male
```

```python
df[df["Team"] == "Finance"] # getting all employees in Finance
```

```python
mask = df["Team"] == "Finance" #mask is storing the boolean series generated by this code
df[mask] # shows us all employees in finance 
```

```python
df[df["Senior Management"]] # getting all the rows where senior mgmt. = True , this is a way of dpoing it
```

```python
df[df["Team"] != "Marketing"] # getting all employees not in marketing
```

```python
#all employees with salary >110k
df[df["Salary"] > 110000]
```

```python
#all employees with bonus% < 1.5%
df[df["Bonus %"] < 1.5]
```

```python
#all employees who started on or before 1 Jan 1985
mask = df["Start Date"] <= "1985-01-01"
df[mask]
```

# Filter with more than one condition (AND)

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Gender"] == "Male"
mask2 = df["Team"] == "Marketing"

df[mask1 & mask2] #extracted a subset of df using 2 conditions that we want both to be true i.e
#male and marketing
```

# Filter with More than One Condition (OR)

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2]
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]
```

# The .isin( ) Method

```python
# .isin () useful for checking multiple methods
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df ["Team"] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3]
```

```python
mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask]
```

# The .between( ) Method

```python
#finding values that are between a given range or date : .between() method
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
#extracting values where salary is between 60k and 70k and both these values are inclusive
df[df["Salary"].between(60000, 70000)]
```

```python
df[df["Bonus %"].between(2.0,5.0)] #finding employees with bonus between 2%-5% and inclusive
```

```python
df[df["Start Date"].between("1991-01-01", "1992-01-01")]
```

```python
df[df["Last Login Time"].between("08:30am", "12:00pm")]
```

# The .duplicated( ) Method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
df[df["First Name"].duplicated()]
```

```python
df[df["First Name"].duplicated(keep = "last")]
#by default the keep parameter = first i.e. this method wont count the first value and then 
#take all the duplicates that follow thereafter. By setting it to last, it is not gonna count 
#the last value in this case
```

```python
df[df["First Name"].duplicated(keep = False)]
#False will mark all the duplicates and extract all the duplicates
```

```python
#if you wanna extract all the unique names and dont want the duplicates
#so put a specuial symbol ~ (tilda) that will reverse the Trues to falses and vice versa
mask = ~df["First Name"].duplicated(keep =False )
df[mask]
```

# The drop.duplicates ( ) Method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
len(df)
```

```python
len(df.drop_duplicates())
#no 2 rows are totally identical and hence no row will be completely dropped
```

```python
df.drop_duplicates(subset = ["First Name"], keep = "first")
```

```python
df.drop_duplicates(subset = ["First Name"], keep = False)
#to remove duplicate in first name col. period, it will remove all the names that occur more than once
```

```python
df.drop_duplicates(subset = ["Team"], keep = False)
#empty dataset coz there is no unique Team. Teams occur multiple times
```

```python
df.drop_duplicates(subset = ["Team", "First Name"], inplace = True)
```

```python
len(df)
```
