# Drop Rows with Null Values

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.tail(3)
```

```python
nba.dropna() # removes any row from a dataframe that has even 1 null value
#the default how parameter is set to 'any'
```

```python
nba.dropna(how = "all") #only drops those rows where all values are None
#however this is not a permanent removal
```

```python
nba.dropna(how = "all", inplace = True) # inpace parameter removes it permanently
```

```python
nba.tail(3)
```

```python
nba.head(3)
```

```python
#can use the subset parameter where we can specify the list of columns where we wanna check for null values
nba.dropna(subset = ["Salary"]) #will drop all rows where salary = nAN
```

# Sort a DataFrame with the .sort_values( ) Method, Part 1 

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
# sorting a dataframe bit complex as pandas doesnt know which col you wanna sort
nba.sort_values("Name", ascending = False)
```

```python
nba.sort_values("Age")
```

```python
nba.sort_values("Salary", ascending = False, inplace = True) # getting the top earners in the NBA
nba.head(3)
```

```python
nba.sort_values("Salary", ascending = False, na_position = "first")
#giving all the null salaries first
```

```python
nba.sort_values("Salary", ascending = False, na_position = "first").tail()
```

# Sort a DataFrame with the sort_values( ) Method, Part II

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Team", "Name"], ascending = False)
#sorting the list first by Team and then by name in desceding order
```

```python
nba.sort_values(["Team", "Name"], ascending = [True, False], inplace = True)
nba.head(3)
# team will be sorted in ascending order, then name in Desc. This is by creating this boolean list
# by calling the inplace parameter we are modifying the original dataframe
# thus can sort multiple columns in a dataframe
```

# Sort DataFrame with the .sort_index( ) Method

```python
nba = pd.read_csv("nba.csv")
nba.head(3)
```

```python
nba.sort_values(["Number", "Salary", "Name"], inplace = True)
nba.tail(3)
```

```python
nba.sort_index(ascending = False, inplace = True)
nba.head(3)
#index sorted in descending order and making the change permanent by using the inplace parameter
```

# Rank Values with the .rank( ) Method

```python
# first removing all the null values to use the rank() method
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
#giving player with highest salary , a rank = 1
# converting floating point to int by astype() method
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending = False)
```
