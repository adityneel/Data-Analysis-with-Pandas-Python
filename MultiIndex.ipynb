# Extract Rows from a MultiIndex Dataframe

```python
bigmac = pd.read_csv("bigmac.csv", parse_dates = ["Date"], index_col = ["Date", "Country"])
bigmac.sort_index(inplace = True)
#sorting the index greatly increases the speed and efficiency
bigmac.head(3)
```

```python
bigmac.loc[("2010-01-01")]
#2010-01-01 will be the only value selected from the outer index and the remaining values 
#from the country index will still be kept
```

```python
bigmac.loc[("2010-01-01", "Brazil")]
#series where index label rep the columns and the value on those columns
#here we only have the price that gets extracted
```

```python
bigmac.loc[("2010-01-01", "Brazil"), "Price in US Dollars"]
#passing a tuple to pass 2 or multiple values to pass horizontally to Pandas
#speicfy it as a series after the tuple parentheis 
#unfortunately in this case there is only one col i.e. for the Price in US dOLLARS
```

```python
bigmac.loc[("2015-07-01", "Chile"), "Price in US Dollars"]
```

# The .pivot( ) Method

```python
sales = pd.read_csv("salesmen.csv", parse_dates = ["Date"])
sales["Salesman"] = sales["Salesman"].astype("category")
sales.head(3)
```

```python
sales["Salesman"].value_counts()
```

```python
sales.pivot(index = "Date", columns = "Salesman", values = "Revenue")
#here choose what columns from original df you wanna serve as new col e.g. Salesman is the new column we chose
#values you choose the column you want the data to be aggregated by and here we chose to aggregate by revenues
```

# The .pivot_table( ) Method

```python
foods = pd.read_csv("foods.csv")
foods.head(3)
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "mean")
```

```python
foods.pivot_table(values = "Spend", index = "Gender", aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "sum")
#performed an aggregation of the spend based on values in city column
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = ["Frequency","City"], aggfunc = "sum")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "count")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "max")
```

```python
foods.pivot_table(values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "min")
```

```python
pd.pivot_table(data = foods, values = "Spend", index = ["Gender", "Item"], columns = "City", aggfunc = "min")
#can also call the dataframe via the pd library as well
```
